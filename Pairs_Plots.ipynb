# Python Pair Plots
## Import Libraries
Installation: `pip install seaborn` or `conda install seaborn`

```python
import seaborn as sns
%matplotlib inline
```

## Import Dataset

```python
tips_econ = sns.load_dataset('tips')
tips_econ.head()
```

## Seaborn Library
### Pairs Plots

```python
sns.pairplot(tips_econ)
```

### Pairs Plot - Separation of Bivariate Conditional Variable

```python
sns.pairplot(tips_econ,hue='sex',palette='coolwarm')
```

### Iris Data

```python
iris_data = sns.load_dataset('iris')
iris_data.head()
```

```python
sns.pairplot(iris_data,hue='species',palette='rainbow')
```
