# Python Swarm Plots
## Import Libraries
Installation: `pip install seaborn` or `conda install seaborn`

```python
import seaborn as sns
%matplotlib inline
```

## Import Dataset

```python
tips_econ = sns.load_dataset('tips')
tips_econ.head()
```

## Swarm Plot

```python
sns.swarmplot(x="day", y="total_bill", data=tips_econ)
```

### Swarm Plot - Separate variable categories

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips_econ, palette="Set1", dodge=True)
```
