## Import Dataset

```python
tips_econ = sns.load_dataset('tips')
tips_econ.head()
```

```python
sns.rugplot(tips_econ['total_bill'])
sns.kdeplot(tips_econ['total_bill'])
```

```python
sns.kdeplot(tips_econ['tip'])
sns.rugplot(tips_econ['tip'])
```
