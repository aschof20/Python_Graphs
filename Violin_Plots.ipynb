# Python Violin Plots
## Import Libraries
Installation: `pip install seaborn` or `conda install seaborn`

```python
import seaborn as sns
%matplotlib inline
```

## Import Dataset

```python
tips_econ = sns.load_dataset('tips')
tips_econ.head()
```

## Violin plots

```python
sns.violinplot(x="day", y="total_bill", data=tips_econ,palette='rainbow')
```

## Violin Plot - comparison of groups

```python
sns.violinplot(x="day", y="total_bill", data=tips_econ,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips_econ,hue='sex',split=True,palette='Set1')
```
