## Clustermap Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
flight_data = sns.load_dataset('flights')
```

```python
flight_data.head()
```

### Clustermap

```python
flight_pivot_table = flight_data.pivot_table(values='passengers',index='month',columns='year')
flight_pivot_table
```

```python
sns.clustermap(flight_pivot_table)
```

```python
sns.clustermap(flight_pivot_table, cmap='coolwarm', standard_scale=1)
```
