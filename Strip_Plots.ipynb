# Python Strip Plots
## Import Libraries
Installation: `pip install seaborn` or `conda install seaborn`

```python
import seaborn as sns
%matplotlib inline
```

## Import Dataset

```python
tips_econ = sns.load_dataset('tips')
tips_econ.head()
```

## Strip Plot

```python
sns.stripplot(x="day", y="total_bill", data=tips_econ, jitter=False)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips_econ)
```

### Strip Plot - separating Variable Categories

```python
sns.stripplot(x="day", y="total_bill", data=tips_econ,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips_econ,hue='sex',palette='Set1',dodge=True)
```
