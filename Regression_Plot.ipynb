# Python Regression Plots
## Install libraries

```python
import seaborn as sns
%matplotlib inline
```

```python
tips_econ = sns.load_dataset('tips')
tips_econ.head()
```

## Linear Model (lm) Plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips_econ)
```

### lm Plot - variable category separation

```python
sns.lmplot(x='total_bill',y='tip',data=tips_econ,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips_econ,hue='sex', palette='coolwarm')
```

### Modifying marker types

```python
sns.lmplot(x='total_bill',y='tip',data=tips_econ,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

### Separating categories into different plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips_econ,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips_econ)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips_econ,col='day',hue='sex',palette='coolwarm')
```

### Size and Aspect Ratio

```python
sns.lmplot(x='total_bill',y='tip',data=tips_econ,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,height=8)
```
