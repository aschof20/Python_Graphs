# Python Joint Plots
## Import Libraries
Installation: `pip install seaborn` or `conda install seaborn`

```python
import seaborn as sns
%matplotlib inline
```

## Import Dataset

```python
tips_econ = sns.load_dataset('tips')
tips_econ.head()
```

## Seaborn Library
### Joint Plots

```python
sns.jointplot(x='total_bill',y='tip',data=tips_econ, kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips_econ, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips_econ, kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips_econ, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips_econ, kind='resid')
```
